## Flowchart

- **Flowchart** คือเครื่องมือที่ช่วยในการอธิบายลำดับความคิดและแสดงตรรกะของผู้เขียนโปรแกรมต่อผู้อ่านโปรแกรม
- มีลักษณะเป็นแผนภาพแสดงการทำงานของโปรแกรม
 - แสดงลำดับ (Sequence)
 - แสดงทางเลือก (Selection)
 - แสดงการทำซ้ำ (Iteration)
- **Flowchart** เปรียบเสมือนการเขียนแบบแปลนของอาคาร ก่อนที่จะลงมือก่อสร้าง
- ในการพัฒนาโปรแกรมควรจะมี Flowchart ก่อนลงมือเขียนโปรแกรม โดยเฉพาะ เมื่อโปรแกรมมีความซับซ้อน หรือมีผู้ร่วมพัฒนาหลายคน

## การเลือก และการสร้างเงื่อนไข

- โดยทั่วไปชุดคำสั่ง (Statement) ในโปรแกรมจะทำงานเรียงตาม ลำดับก่อนหลัง (Sequence) จากโค้ดบรรทัดบนลงไปล่าง
- แต่บางครั้งการเขียนโปรแกรมจำเป็นต้องมี ทางเลือก (Selection) เช่น ในตัวอย่าง 1 โดยอาศัย เงื่อนไข (Condition) ในการพิจารณา
- **เงื่อนไข**จะถูกสร้างขึ้นโดยอาศัยรูปแบบคำสั่ง 2 ลักษณะต่อไปนี้ประกอบกัน
 - **Relational Expression** - การเปรียบเทียบ
 - **Logical Expression** - การคำนวณทางตรรกะ



### Relational Expression

การประเมิน (Evaluate) ความสัมพันธ์เชิงเปรียบเทียบ (comparison) ว่าเป็นไปตามที่กำหนดหรือไม่

การเปรียบเทียบมักจะอาศัย Relational Operators ต่อไปนี้

| Relational Operator | ความหมาย            |
|:-------------------:|:--------------------|
|         `<`         | น้อยกว่า            |            
|         `>`         |  มากกว่า            |             
|        `<=`         | น้อยกว่าหรือเท่ากับ | 
|        `>=`         | มากกว่าหรือเท่ากับ  |
|        `==`         | เท่ากับ             |
|        `!=`         | ไม่เท่ากับ          |

### การใช้ **in** และ **not in** Operator

เราสามารถใช้ **`in`** และ **`not in`** ในการตรวจสอบความเป็นสมาชิก (membership test) ของข้อมูลว่าอยู่ภายใน collection ที่เราสนได้ เช่น

```python
>>> 5 in [2,3,4,5,9,7]
True

>>> "Ann" in ["John","James","Ken","Annie"]
False

>>> 10 in (5,10,15,20,25,30)
True

>>> "d" in "Dome Potikanond"
True
```

### Logical Expression

| Logical Operator | ความหมาย                                   |
|:----------------:|:-------------------------------------------|
|      `not`       | ค่าตรงกันข้าม                              |

|       `or`       | เป็นจริง ถ้าอย่างน้อยค่าใดค่าหนึ่งเป็นจริง |

### Example 3 : if (Compound Statement)

- จากตัวอย่างที่ 2 เราสามารถแก้ไขส่วนของโปรแกรม (if-statement)
- หากข้อมูลมีค่าน้อยกว่าศูนย์ ให้คำนวณค่าสัมบูรณ์ และ**เพิ่มค่าดังกล่าวขึ้นไปเท่ากับ 10**

```python

number = int(input("Enter your number:"))
if number < 0:
   number = -number
   number = number + 10
print("The absolute value is", number)  # next-statement
```

```python

Blackpinks = ["Jisoo","Rose","Jennie","Lisa"]
name = input()

# ตรวจสอบความเป็นสมาชิก และแสดงผลลัพธ์
if name in Blackpinks:
  print(f'{name} is a Blackpink member.')
  print(f'{name} is number {Blackpinks.index(name) + 1} in Blackpink.')
```

### Example 4 : if-else

- โปรแกรมคำนวณการแปลงอุณหภูมิในแบบเซลเซียสและแบบฟาเรนไฮท์

```python 

temp = float(input("Enter temperature: "))
temp_type = input("Enter f (Fahrenheit) or c (Celsius) for unit: ")

# เกิดอะไรขึ้นหาก temp_type มีค่าเป็น "F" ?

   c = (5.0/9) * (temp - 32)

else:
   f = (9.0/5) * temp + 32
   print(f"The equivalent Fahrenheit temp is {f:.2f}")

```


### Example 6 : Nested if

- โปรแกรมรับค่าเลขจำนวนเต็ม 2 จำนวน (คั่นด้วยช่องว่าง)
- ตรวจสอบว่า **`เลขตัวแรก`** ว่าอยู่ในช่วง `0-9` และ ตรวจ **`เลขตัวที่สอง`** ว่าอยู่ในช่วง `10-99` หรือไม่
- แสดงผลการตรวจสอบดังกล่าวบนหน้าจอ

```{code-block} python

print("Check if num1 is in [0-9] and num2 is in [10-99].")
num1,num2 = input("num1 and num2: ").split()
num1=int(num1); num2=int(num2)

if num1 >= 0 and num1 <= 9:
   if num2 >= 10 and num2 <= 99:
      print("Both numbers are entered correctly.")
   else:
      print("Only num1 is entered correctly.")

else:
   if num2 >= 10 and num2 <= 99:
      print("Only num2 is entered correctly.")
   else:
      print("Both numbers are NOT entered correctly.")
```

## Summary

- **Flowchart** คือผังภาพแสดงงานและการทำงานของโปรแกรม
 - อาจเป็นแบบ**ลำดับ** (Sequence), **ทางเลือก** (Selection), หรือ **ทำซ้ำ** (Iteration)
 - มีประโยชน์ในการทำความเข้าใจการทำงานของโปรแกรมที่มีความซับซ้อน
 - ใช้ตรวจสอบการทำงานก่อนทำการลงมือเขียนโปรแกรมจริง
- การเขียนโปรแกรมแบบมีทางเลือก ทำได้โดยการระบุ**เงื่อนไข** ที่ต้องมีการทดสอบ ก่อนทำตามทางที่ได้เลือก
 - ระบุคำสั่งที่ต้องการให้ถูกกระทำเมือเงื่อนไขเป็น **True** หรือ **False**
- **เงื่อนไข (Condition)** คือรูปแบบ (expression) ที่จะถูกทดสอบเพื่อการตัดสินใจ แบ่งเป็น Relational expression และ Logical expression

- คำสั่งในการสร้างทางเลือก ได้แก่
 - if
 - if–else
 - if–else if (elif)
